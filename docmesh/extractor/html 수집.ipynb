```python
from IPython.display import clear_output
import matplotlib.pyplot as plt
import time
```

```python
import sys
sys.path.append("../../")
```

```python
# 구 버전 문서 a제외
excluded_patterns = [r"^https://python\.langchain\.com/v\d+\.\d+/"]
```

```python
from docmesh.extractor import *

start_url = "https://docs.smith.langchain.com/"

crawler = FileSavingCrawler(start_url, 
                            save_path="./langsmith",
                            excluded_patterns=excluded_patterns, use_max_pages=False)
```

```python
num_batch = 10

num_done = 0
dt_avg = 0
list_num_to_visit = [0]
```

```python
while True:
    t_s = time.time()
    clear_output(wait=True)
    results = crawler.crawl(num_batch=num_batch)
    num_done += 1
    t_e = time.time()
    
    dt = t_e - t_s
    dt_avg = (dt_avg * (num_done-1) + dt) / num_done
    t_remain_expected = dt_avg * len(crawler.to_visit) / num_batch / 60
    
    list_url = ""
    for url in crawler.results:
        list_url += f"{url}\n"
        
    print(
        f"""
{num_done}번째 배치 완료
t_remain_expected(min) : {t_remain_expected}

{len(crawler.to_visit)}
{len(crawler.visited)}
{list_url}
    """
         )

    if len(crawler.to_visit) == 0:
        break
```
